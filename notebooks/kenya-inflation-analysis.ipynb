```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/cleaned/kenya_final.csv")
print(df.shape)
print(df.head())
```

```python
df.describe().round(2)
```

```python
fig, ax = plt.subplots (figsize=(12, 5))
ax.plot(df["year"], df["inflation"], color="steelblue", linewidth=2)
ax.set_title("Kenya Inflation Rate (1980-2024)")
ax.set_xlabel("Year")
ax.set_ylabel("Inflation (%)")
ax.axhline(y=5, color="red",linestyle="--", linewidth=1, label="5% target")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(df["year"], df["inflation"], color="steelblue",linewidth=2)
ax.set_title("Kenya Inflation Rate (1980-2024)")
ax.set_xlabel("Year")
ax.set_ylabel("Inflation (%)")
ax.axhline(y=5, color="red",linestyle="--",linewidth=1, label="5% target")
ax.annotate("Structural adjustment",
            xy=(1993, 45.98), 
            xytext=(1996, 42),
            arrowprops=dict(arrowstyle="->",color="black"), 
            fontsize=9)
ax.annotate("Global food crisis",xy=(2008, 26.24),
            xytext=(2010, 30),
            arrowprops=dict(arrowstyle="->",color="black"), 
            fontsize=9)
ax.annotate("COVID-19",xy=(2020, 5.40),
            xytext=(2017, 2),
            arrowprops=dict(arrowstyle="->",color="black"), 
            fontsize=9)
ax.legend()
plt.tight_layout()
plt.show()
            
```

```python
fig, axes = plt.subplots(3, 2,figsize=(14, 10))
axes[0, 0].plot(df["year"], df["inflation"], color="steelblue",linewidth=2)
axes[0, 0].set_title("Inflation (%)")
axes[0, 0].set_xlabel("Year")

axes[0, 1].plot(df["year"], df["exchange_rate"], color="darkorange",linewidth=2)
axes[0, 1].set_title("Exchange Rate (KES/USD)"), 
axes[0, 1].set_xlabel("Year")

axes[1, 0].plot(df["year"], df["broad_money"], color="green",linewidth=2)
axes[1, 0].set_title("Broad Money (% of GDP)")
axes[1, 0].set_xlabel("Year")

axes[1, 1].plot(df["year"], df["gdp_growth"], color="purple",linewidth=2)
axes[1, 1].set_title("GDP Growth (%)")
axes[1, 1].set_xlabel("Year")

axes[2, 0].plot(df["year"], df["food_index"], color="brown",linewidth=2)
axes[2, 0].set_title("Food Production Index")
axes[2, 0].set_xlabel("Year")

axes[2, 1].set_visible(False)

plt.suptitle("Kenya Economic Indicators (1980-2024)", fontsize=14)
plt.tight_layout()
plt.show()

                     
```

```python
numeric_cols = ["inflation", "exchange_rate", "broad_money", "gdp_growth", "food_index"]
corr = df[numeric_cols].corr()
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
ax.set_title("Correlation Matrix Of Kenya Economic Indicators")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df["exchange_rate"], df["inflation"], color="steelblue", alpha=0.7)
ax.set_title("Inflation vs Exchange Rate (1980-2024)")
ax.set_xlabel("Exchange Rate (KES/USD)")
ax.set_ylabel("Inflation (%)")
plt.tight_layout()
plt.show()
                       
                       
```
